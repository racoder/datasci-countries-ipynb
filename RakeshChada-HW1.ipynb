#<h3>Extracting the CSV File's content into a pandas data frame

```python
df = DataFrame(pd.read_csv('http://www.cs.stonybrook.edu/~skiena/591/hw1/country-data.csv'))
```

#<h3>Data Wrangling
<ul><li>Entries with missing values ("none", "unknown" etc) are replaced with numpy.Nan</li><li>Latitude and Longitude are converted from degree,minutes,seconds format to floating point number format</li><li>Missing Values (NaN entries) are filled using interpolation</li></ul></h3>

```python
df = df.replace('unknown', np.NaN)
df = df.replace('none', np.NAN)
df = df.replace('NA', np.NAN)
#Replace the Military Expenditure/GDP entry with it's right representation 
df = df.replace('1.78%','1.78')
```

```python
def latLongToFloat(latLongValue):
    if latLongValue is np.NaN:
        return 0
    direction = {'N':1, 'S':-1, 'E': 1, 'W':-1}
    new = latLongValue.split()
    new_dir = new.pop()
    new.extend([0,0,0])
    return (int(new[0])+int(new[1])/60.0+int(new[2])/3600.0) * direction[new_dir]
```

```python
df['Latitude of Capital'] = df['Latitude of Capital'].apply(latLongToFloat)

df['Longitude of Capital'] = df['Longitude of Capital'].apply(latLongToFloat)

floatDf = df.iloc[:,5:21].astype(float)

finalDf = floatDf.interpolate()
```

#<h3>Data Map for Literacy Rate</h3>

```python
lats, lons = finalDf['Latitude of Capital'], finalDf['Longitude of Capital']
literacyRates = finalDf['Literacy Rate (%)']
labels = df['Country Code']

# --- Build Map ---
from mpl_toolkits.basemap import Basemap

def get_marker_color(literacyRate):
    # Returns green for small literacyRates, yellow for moderate
    #  literacyRates, and red for significant literacyRates`.
    if literacyRate < 50.0:
        return ('r')
    elif literacyRate < 75.0:
        return ('y')
    elif literacyRate < 90.0:
        return ('b')
    else:
        return ('g')
 
map = Basemap(projection='robin', resolution = 'l', area_thresh = 1000.0,
              lat_0=0, lon_0=0)
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color = 'gray')
map.drawmapboundary()
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))
 
min_marker_size = 2.5
plotLabelDict = {'r':"Literacy Rate < 50%",'y':"Literacy Rate between 50%-75%",'b':"Literacy Rate between 75%-90%",'g':"Literacy Rate > 90%"}

for lon, lat, ltrate, countryLabel in zip(lons, lats, literacyRates, labels):
    x,y = map(lon, lat)
    msize = (ltrate * min_marker_size)/20
    marker_string = get_marker_color(ltrate)
    mapPlt = map.plot(x, y, marker_string + 'o', markersize=msize, label=plotLabelDict[marker_string])


proxy = [plt.Rectangle((0,0),1,1,fc = color) for color in ['r','y','b','g']]

leg = plt.legend(proxy, ["Literacy Rate < 50%", "Literacy Rate between 50%-75%", "Literacy Rate between 75%-90%", "Literacy Rate > 90%"], loc=4)
```

#<h3><b>Representation of the Data Map:</b>

<ul><li>Countries are divided into four categories<ul><li>Ones with literacy rate < 50% (indicated by a red circle)</li><li>Ones with literacy rate between 50% - 75% (indicated by a yellow circle)</li><li>Ones with literacy rate between 75% - 90% (indicated by a blue circle)</li><li>Ones with literacy rate greater than 90% (indicated by a green circle)</li></ul></li><li>Size of the circle is proportional to the literacy rate</li></ul>

#<h3>Date of Founding/Independence vs Population</h3>


```python
#extract the 2 columns
dateDf = df[["Date of Founding/Independence", "Population"]] 

#filter rows containing NaN in any of the columns
dateDf = dateDf.dropna(subset = ['Date of Founding/Independence', 'Population']) 

dateList=[]

for d in dateDf["Date of Founding/Independence"]:
    try:
        dateFormat = dt.datetime.strptime(d,'%m/%d/%Y')
        dateList.append(date2num(dateFormat.date()))
    except ValueError:
        #This is the case where date is mentioned in yyyy format
        try:
            dateFormat = dt.datetime.strptime(d,'%Y')
            d = "1/1/" + d
            newDateFormat = dt.datetime.strptime(d,'%m/%d/%Y')
            dateList.append(date2num(newDateFormat.date()))
        except ValueError:
            #This is the case of a 3 digit year
            d = "0" + d
            dateFormat = dt.datetime.strptime(d,'%Y')
            d = "1/1/" + d
            newDateFormat = dt.datetime.strptime(d,'%m/%d/%Y')
            dateList.append(date2num(newDateFormat.date()))

populationList = [int(p)/10000 for p in dateDf["Population"].astype(float)]

points = zip(dateList,populationList)
sorted_points = sorted(points)
sortedDates = [point[0] for point in sorted_points]
sortedPopulation = [point[1] for point in sorted_points]

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.plot(sortedDates,sortedPopulation,'o-')
plt.xlabel('Date of Founding/Independence',fontsize=16)
plt.ylabel('Population in multiples of 10000',fontsize=16)
plt.xlim((352000.0))
plt.gcf().autofmt_xdate()
```

#<h3>Correlation
A correlation co-efficient value table is constructed between all pairs of variables.</h3>

```python
slicedDf = finalDf.iloc[:,2:]
#Replacing GDP column with GDP Per Capita
slicedDf['GDP (PPP) in US $'] = slicedDf['GDP (PPP) in US $']/slicedDf['Population']

corrDf = slicedDf.corr()
corrDf
```

#<h3>
Variables that correlate most and those that correlate least are printed out. 

 - While measuring the strength of the correlation, negative values are interpreted as positive values as strength of the correlation is independent of the sign.

 - One interesting phenomenon that can be observed is that correlation is not symmetric between variables. For example, Internet Users(circa 2009) is most correlated with Labor Force whereas Labor Force is most correlated with Population.</h3>

```python
#replace all -ve values with +ve values
corrDf=corrDf.where(corrDf>0, corrDf * -1)

maxMinCorrDf = pd.DataFrame()

columnNames = corrDf.columns
i=0
for colName in columnNames:
    slicedCorrDf = corrDf[colName].iloc[:i]
    if i != len(columnNames) - 1:
        slicedCorrDf = slicedCorrDf.append(corrDf[colName].iloc[i+1:])
    maxCorr = pd.Series(slicedCorrDf.idxmax(axis=0),name='Most Correlated Variable', index=[colName])
    minCorr = pd.Series(slicedCorrDf.idxmin(axis=0),name='Least Correlated Variable',index=[colName])
    maxMinCorrDf = maxMinCorrDf.append(pd.concat([maxCorr,minCorr],axis=1))
    i = i + 1

maxMinCorrDf
```

#<h3>P-Values

Pearson Permutation Test is used in computing the P-Value for each pair. 

</h3>

```python
from scipy.stats.stats import pearsonr

columnNames = slicedDf.columns

columnsToCompare = {}

pValueDict = {}

for i in range(0,len(columnNames)):
    columnsToCompare[i] = range(0,len(columnNames))
    
pdDataList = []

for i in range(0,len(columnNames)):
    pValueList = []
    for columnIndex in columnsToCompare[i]:
        corrxy,pxy = pearsonr(slicedDf[columnNames[i]],slicedDf[columnNames[columnIndex]])
        pValueList.append(pxy)
        pValueDict[columnNames[i]] = pValueList

for i in range(0,len(columnNames)):
    pdDataList.append(pValueDict[columnNames[i]])

pdData = np.array(pdDataList)
pValueDf = pd.DataFrame(data = pdData, index=columnNames, columns=columnNames)
pValueDf
```

#<h3>Analysis:
   - Population: Least correlated variable with population is Net Migration Rate but their p-value is 0.94. So, there's a very high probability that these are related thereby contradicting what the pearson correlation co-efficient says. Similarly, least correlated variable with Labor Force is Net Migration Rate but their p-value is 0.91. So, there's a good probability that they are related
   - Life Expectancy: Most correlated variable is Literacy Rate and low p-value confirms it. Least correlated is Population and high p-value implies that might not actually be true
   - GDP (Per Capita): Most correlated variable is Life Expectancy and low p-value confirms it. Least correlated is Renewable Water Resources and high p-value implies that might not actually be true

#<h3>Linear Regression

A linear regression model for GDP is built by taking 5 variables (Life Expectancy', 'Literacy Rate (%)', 'Commercial Prime Lending Rate', 'Net Migration Rate/1000 Population). These five variables are picked on the basis of their p-values and correlation co-efficients calculated above. They have either high correlation co-efficient with low p-values or high p-values with low correlation co-efficient. The dataset is split into half to train the model and the resulting model is used for prediction of the GDP.</h3>

```python
import sklearn
from sklearn import linear_model
import statsmodels.formula.api as sm

clf = linear_model.LinearRegression()

x = slicedDf[['Life Expectancy', 'Literacy Rate (%)', 'Commercial Prime Lending Rate', 'Net Migration Rate/1000 Population']]
y = slicedDf['GDP (PPP) in US $']

features_train, features_test, target_train, target_test = sklearn.cross_validation.train_test_split(x,y, test_size=0.5)
clf.fit(features_train,target_train)
print "Co-effecients are",clf.coef_.astype(int),"Intercept is", clf.intercept_.astype(int)
gdp_predicted = clf.predict(x)
newDf = pd.concat([pd.Series(gdp_predicted),y],keys=['Predicted GDP','Actual GDP'],axis=1)
newDf = newDf.set_index(df['Country Name'])
newDf['Difference from Predicted Value'] = newDf['Actual GDP'] - newDf['Predicted GDP']
newDf = newDf.sort(columns='Difference from Predicted Value')
newDf
```

#Problem 4 - Ranking Function for Social Welfare

Variables selected to measure Social Welfare are GDP (PPP) in US $ (Per Capita), Life Expectancy, Literacy Rate (%), Health Expenditure/GDP and Internet Users(circa 2009).

A rise in per capita GDP signals growth in the economy. Higher values of Life Expectancy, Literacy Rate and Health Expenditure/GDP would definitely imply a better social welfare of the people in the country. Hence, these variables are used to represent the rank of a country. Similarly, a greater number of internet users might indicate a better economic situation of the country and hence it can be a substantial variable that measures the rank of a country.

GDP(Per Capita) and Internet Users are scaled down by a factor of 10000 so that their values lie in the same range as that of other variables involved in the measure of social welfare.

Then, values of each of these five variables are added to represent a rank metric for the country. This sum value is then ranked using the dataframe rank() function.

All countries with their ranks in sorted order are displayed below.

```python
rankColumns=pd.concat([slicedDf['GDP (PPP) in US $'],slicedDf['Life Expectancy'],slicedDf['Literacy Rate (%)'],slicedDf['Health Expenditure/GDP'],slicedDf['Labor Force']],axis=1,
                      keys=['Rank - GDP (PPP) in US $','Rank - Life Expectancy','Rank - Literacy Rate (%)','Rank - Health Expenditure/GDP','Rank - Labor Force'])
rankDf = pd.DataFrame(slicedDf['GDP (PPP) in US $']/10000 + slicedDf['Life Expectancy'] + slicedDf['Literacy Rate (%)'] + slicedDf['Health Expenditure/GDP'] + slicedDf['Internet Users (circa 2009)']/10000, columns=['Overall Rank'], index=slicedDf.index)
rankDf.index = df['Country Name']
rankDf = rankDf['Overall Rank'].rank(ascending=False)
rankDf.sort()
rankDf
```

These results obtained from the data look convincing when compared with the actual social welfare of the countries. Each variable is believed to have a strong impact on the social welfare of a country and hence they are appropriately ranked lower. #Problem 5 - Distance Function between countries

  - Variables used for distance measure are 'Population','Life Expectancy', 'GDP (PPP) in US $', 'Literacy Rate (%)', 'Net Migration Rate/1000 Population', 'Labor Force', 'Internet Users (circa 2009)'
  - Distance Function Used is Euclidean Distance
  - Table of distances between each pair of country is displayed below using above variables and the distance function

```python
from scipy.spatial.distance import pdist, squareform
scaledDf = slicedDf/10000
distanceDf = scaledDf[['Population','Life Expectancy', 'GDP (PPP) in US $', 'Literacy Rate (%)', 'Net Migration Rate/1000 Population', 'Labor Force', 'Internet Users (circa 2009)']]
Y = pd.DataFrame(squareform(pdist(distanceDf, 'euclidean')))
Y.columns = df['Country Name']
Y.index = df['Country Name']
Y
```

# Nearest and Farthest neighbours

```python
maxY = pd.Series(Y.idxmax(axis=0),name='Farthest neighbour')

minY = Y.replace(0.000000,158978.429497)
minY = pd.Series(minY.idxmin(axis=0),name='Nearest neighbour')

neighbourDf = pd.concat([maxY,minY],axis=1)

neighbourDf
```

This distance function is reasonably good in estimating similar countries. For example, one expects India and China to be close neighbours. However, this similarity is restricted to a few factors like Population, Life Expectancy, GDP, Internet Users etc. Assuming that the selected variables, as a set, represent social welfare, it can be said that the distance function does a good job in estimating countries that have similar quality of social welfare.
