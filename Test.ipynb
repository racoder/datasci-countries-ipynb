#Conversion of latitude,longitude (from degree,minutes,seconds format to floating point number)

```python
def latLongToFloat(latLongValue):
    #TODO:handle NaNs in pairs when plotting later
    if latLongValue is np.NaN:
        return 0
    direction = {'N':1, 'S':-1, 'E': 1, 'W':-1}
    new = latLongValue.split()
    new_dir = new.pop()
    new.extend([0,0,0])
    return (int(new[0])+int(new[1])/60.0+int(new[2])/3600.0) * direction[new_dir]

lat, lon = '38 53 N', '77 02 W'
print latLongToFloat(lat), latLongToFloat(lon)
```
